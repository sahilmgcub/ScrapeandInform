```python
from tkinter import *
import threading
root = Tk()
root.title("IPL Season 13 Live Score")

url = 'https://www.cricbuzz.com/'
import requests
from bs4 import BeautifulSoup

def Fetch_score(livescore):
    Firstteam, Secondteam, Firstteam_score, Secondteam_score, results = livescore
    cricbuzzpage = requests.get(url)
    soup = BeautifulSoup(cricbuzzpage.text, 'html.parser')
    findteam1 = soup.find_all(class_="cb-ovr-flo cb-hmscg-tm-nm")[0].get_text()
    findteam2 = soup.find_all(class_="cb-ovr-flo cb-hmscg-tm-nm")[1].get_text()

    findteam1_score = soup.find_all(class_="cb-ovr-flo")[8].get_text()
    findteam2_score = soup.find_all(class_="cb-ovr-flo")[10].get_text()
    findresult = soup.find(class_="cb-ovr-flo cb-text-complete").get_text()
    
    Firstteam.config(text=findteam1)
    Secondteam.config(text=findteam2)
    Firstteam_score.config(text=findteam1_score)
    Secondteam_score.config(text=findteam2_score)
    results.config(text=findresult)
    
root.geometry("400x200")
  
 

a = Label(root, text='IPL 2020 Season 13', font=("", 20))
a.grid(row=0,column=1,columnspan=2, padx=5, pady=5)
Firstteam = Label(root, text="Team 1", font=("", 10))
Firstteam.grid(row=1, column=1,padx=5, pady=5)
  
Secondteam = Label(root, text="Team 2", font=("", 10))
Secondteam.grid(row=1, column=3,padx=5, pady=5)

Firstteam_score = Label(root, text="hit refresh", font=("", 10))
Firstteam_score.grid(row=2, column=1,padx=5, pady=5)
Secondteam_score = Label(root, text="hit refresh", font=("", 10))
Secondteam_score.grid(row=2, column=3,padx=5, pady=5)
results = Label(root, text="Results", font=("", 10))
results.grid(row=4,column=3 ,columnspan=1)
livescore = [Firstteam, Secondteam, Firstteam_score, Secondteam_score, results]
refresh = Button(root, text='Refresh',
                     command=lambda: Fetch_score(livescore), height=1, width=10, font=("", 10))
refresh.grid(row=3,column=3,columnspan=1)

root.mainloop()
```
